```python
import tensorflow as tf
import keras_nlp

# Load the trained model
model_path = "/kaggle/input/twitter-disaster-model/keras/default/1/my_model.keras" 
classifier = tf.keras.models.load_model(model_path)
```

```python
def classify_text(input_text):
    """Classifies input text as Disaster or Non-Disaster."""
    if not input_text.strip():
        return "Invalid input. Please enter some text."

    prediction = classifier.predict([input_text]) 

    # Determine class label
    class_label = "Disaster" if prediction[0][1] > prediction[0][0] else "Non-Disaster"
    
    return f"Prediction: {class_label}"
```

```python
#Loop for user input
print("Twitter Disaster Prediction Model (Type 'exit' to quit)")
while True:
    user_input = input("Enter a tweet: ").strip()
    if user_input.lower() == "exit":
        print("Exiting...")
        break
    
    result = classify_text(user_input)
    print(result)
```
